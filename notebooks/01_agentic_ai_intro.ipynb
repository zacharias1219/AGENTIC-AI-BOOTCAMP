### 3. The 'Reasoning' Prompt
We don't just want an answer; we want a plan. We use a **System Prompt** to define the behavior.

```python
model = genai.GenerativeModel('gemini-flash-lite-latest')

goal = "Plan a 3-day trip to Tokyo"

prompt = f"""
You are a travel planner agent.
User Goal: {goal}

Your Task:
1. Break this down into steps.
2. Provide a reasoning for each step.
3. Output a final itinerary.
"""

response = model.generate_content(prompt)
print(response.text)
```
